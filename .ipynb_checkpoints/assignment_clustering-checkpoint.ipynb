**Q1.** This is a question about clustering. We want to investigate how adjusting the "noisiness" of the data impacts the quality of the algorithm and the difficulty of picking $k$.

1. Run the code below, which creates four datasets: `df0_125`, `df0_25`, `df0_5`, `df1_0`, and `df2_0`. Each data set is created by increasing the amount of `noise` (standard deviation) around the cluster centers, from `0.125` to `0.25` to `0.5` to `1.0` to `2.0`.

```
import numpy as np
```

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?
3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$. (Pay attention to the vertical axis across plots, or put all the scree curves on a single canvas.)
4. Explain the intuition of the elbow, using this numerical simulation as an example.

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
````

**1.**

````python
def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df
````

````python
df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

**2.**

````python
sns.scatterplot(x='x1',y='x2',hue='group',data=df0_125).set_title("Scatterplot (Noise = 0.125)")
````

````python
sns.scatterplot(x='x1',y='x2',hue='group',data=df0_25).set_title("Scatterplot (Noise = 0.25)")
````

````python
sns.scatterplot(x='x1',y='x2',hue='group',data=df0_5).set_title("Scatterplot (Noise = 0.5)")
````

````python
sns.scatterplot(x='x1',y='x2',hue='group',data=df1_0).set_title("Scatterplot (Noise = 1.0)")
````

````python
sns.scatterplot(x='x1',y='x2',hue='group',data=df2_0).set_title("Scatterplot (Noise = 2.0)")
````

As the noise goes up from 0.125 to 2.0, the visual distinctiness of the clusters decreases. This means that when the noise increases, the clusters start to overlap on top of each other. It can be seen when the noise is equal to 1.0 the colors begin to spread into other groups.

**3.**

````python
def compute_SSE(df, k_bar=10):
    X = df[['x1','x2']].values
    SSE = []
    for k in range(k_bar):
        model = KMeans(n_clusters=k+1, max_iter=300, n_init=10, random_state=0)
        model.fit(X)
        SSE.append(model.inertia_)
    return SSE

datasets = {
    "Noise=0.125": df0_125,
    "Noise=0.25": df0_25,
    "Noise=0.5": df0_5,
    "Noise=1.0": df1_0,
    "Noise=2.0": df2_0
}

results = {name: compute_SSE(df) for name, df in datasets.items()}

# Plot all scree curves together
plt.figure(figsize=(10,6))
for name, SSE in results.items():
    sns.lineplot(x=np.arange(1, 11), y=SSE, label=name)
````

When the noise is 0.125 and 0.25, the SSE drops sharply from k=1 to k=3 then flattens. This transition creates a clar elbow a k=3, matching the true number of clusters seen in the scatterplots from question 2. When the noise is 0.5, the elbow is still visible, but less sharp in that the drop from k=2 to k=3 can be seen, but there is less definition. When the noise is 1.0, the sharpness at k=3 is very faint and almost looks like a smooth line from k=2 to beyond. Lastly when the noise is 2.0, there are defined drops, but it can be seen that where the elbow seems to have a sharpness is at k=4 rather than k=3.

**4.**

The elbow represents the point where adding more clusters stops providing large improvements in compactness. Initially, increasing k reduces error sharply. However, after the elbow, further reductions in error mostly come from fitting noise rather than meningful groups. In the simulation, it was seen that low noise, 3 natural clusters existed and the elbow was visibly at the right k-value. With high noise, points began to overlap and there was no longer a clear elbow.

**1.**

````python
df = pd.read_csv("./data/2022 election cycle fundraising.csv")
df = df.loc[:, ['Member', 'Raised', 'Spent', 'Cash on Hand', 'Debts']]
df.head()
````

````python
df['Raised'] = pd.to_numeric(df['Raised'].replace(r'[\$,]', '', regex=True))
df['Spent'] = pd.to_numeric(df['Spent'].replace(r'[\$,]', '', regex=True))
df['Cash on Hand'] = pd.to_numeric(df['Cash on Hand'].replace(r'[\$,]', '', regex=True))
df['Debts'] = pd.to_numeric(df['Debts'].replace(r'[\$,]', '', regex=True))
df.head()
````

Cleaned the data by removing '$' and commas within numeric values. Additional handling of DataType from Object to float.

**2.**

````python
# Max-min normalization function
def maxmin(x):
    x = (x-min(x))/(max(x)-min(x))
    return x

# Normalize Raised and Spent
X = df.loc[:, ['Raised', 'Spent']]
X = X.apply(maxmin)

# Scree Plot
SSE = []
k_bar = 11
for k in range(k_bar):
    kmeans = KMeans(n_clusters=k+1, max_iter=300, n_init=10, random_state=0).fit(X)
    SSE.append(kmeans.inertia_)

sns.lineplot(x=np.arange(1, k_bar+1), y=SSE).set_title('Scree Plot (Raised vs Spent)')
````

````python
# Find optimal k value
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
df['cluster_RS'] = kmeans.fit_predict(X)
````

````python
# Scatterplot
sns.scatterplot(x='Raised', y='Spent', hue='cluster_RS', data=df)
````

````python
cluster_counts = df['cluster_RS'].value_counts()
print(cluster_counts)

smallest_clusters = cluster_counts.index[-2:]

df_small = df[df['cluster_RS'].isin(smallest_clusters)][['Member', 'Raised', 'Spent', 'cluster_RS']]
print(df_small)
````

Cluster 0 was scattered around spending and raising little with all points falling under 2 for both spent and raised. The cluster also comprised the most membership. Cluster 1 and 2 had the smallest clusters with cluster 2 being in the middle and cluster 1 having members spend and raise the most. The three candidates that fell into this cluster were Mark Kelly, Raphael Warnock, and Val Demings. From an initial lookup, Mark Kelly ran for senator in Arizona, winning with a 51.39 to 46.51 percentage. The closest of the three high spenders was Raphael Warnock's race where he won 51.40% t 48.60% of the Republican candidate. In Val Demings' case, although she was a high spender, she did not win Florida Senate in 2022.

````python
X = df.loc[:, ['Cash on Hand', 'Debts']]
X = X.apply(maxmin)

# Scree Plot
SSE = []
k_bar = 11
for k in range(k_bar):
    kmeans = KMeans(n_clusters=k+1, max_iter=300, n_init=10, random_state=0).fit(X)
    SSE.append(kmeans.inertia_)

sns.lineplot(x=np.arange(1, k_bar+1), y=SSE).set_title('Scree Plot (Cash on Hand vs Debts)')
````

Within these clusters, David Trone and Ron Johnson were the two that ended with large debts, with Ron Johnson appearing in the smaller clusters in part 2.
